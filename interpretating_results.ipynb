```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Final.csv')
print(df.head())
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
print(df.head())
```

```python
rows=[]
with open('Final.csv', 'r') as f:
    csv_r = csv.reader(f)
    for star in csv_r:
        rows.append(star)
```

```python
headers = rows[0]
star_data = rows[1:]
name = []
distance = []
radius = []
mass = []
gravity = []
```

```python
for star in star_data:
    name.append(star[1])
    distance.append(float(star[2]))
    mass.append(float(star[3]))
    radius.append(float(star[4]))
    gravity.append(float(star[5]))
```

```python
import seaborn as sns

sns.barplot(x=name, y=distance)
plt.xlabel('NAMES')
plt.ylabel('DISTANCE')
plt.show()
```

```python
sns.barplot(x=name, y=mass)
plt.xlabel('NAMES')
plt.ylabel('MASS')
plt.show()
```

```python
sns.barplot(x=name, y=radius)
plt.xlabel('NAMES')
plt.ylabel('RADIUS')
plt.show()
```

```python
sns.barplot(x=name, y=gravity)
plt.xlabel('NAMES')
plt.ylabel('GRAVITY')
plt.show()
```
